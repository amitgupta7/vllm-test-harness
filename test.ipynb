```python
import torch

print(f"PyTorch version: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")

if torch.cuda.is_available():
    num_gpus = torch.cuda.device_count()
    print(f"Number of GPUs detected by PyTorch: {num_gpus}")

    for i in range(num_gpus):
        print(f"--- GPU {i} ---")
        print(f"Name: {torch.cuda.get_device_name(i)}")
        print(f"Total Memory: {torch.cuda.get_device_properties(i).total_memory / 1024**3:.2f} GB")
        # Note: Free memory is not directly available via a simple torch call without more complex allocation
        print(f"Current CUDA Memory Allocated on GPU {i}: {torch.cuda.memory_allocated(i) / 1024**3:.2f} GB")
        print(f"Current CUDA Memory Cached on GPU {i}: {torch.cuda.memory_reserved(i) / 1024**3:.2f} GB") # Total managed memory by PyTorch
else:
    print("CUDA is not available. vLLM will not work.")
```

```python
try:
    import vllm
    print(f"vLLM version: {vllm.__version__}")
    print("vLLM is installed.")
except ImportError:
    print("vLLM is NOT installed. vLLM serving cannot proceed.")

# Final summary
print("\n--- Health Check Summary ---")
if torch.cuda.is_available() and torch.cuda.device_count() >= 2:
    print(f"✅ PyTorch sees {torch.cuda.device_count()} GPU(s).")
    print("✅ Basic memory and operation tests completed.")
    print("ℹ️  Ensure vLLM configuration (e.g., `tensor_parallel_size`) matches the number of GPUs for optimal performance.")
else:
    print("❌ Issue: PyTorch does not detect 2 GPUs or CUDA is not available.")
print("--- End of Health Check ---")
```
